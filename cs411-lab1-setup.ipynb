```python
class TreeNode:
    def __init__(self, value = None):
        self.value = value
        self.children = []
    
    def addChild(self, child):
        self.children.append(child)

def printTree(node, level=0):
    space = " " * level
    print(f"{space}|-- {node.value}")
    for child in node.children:
        printTree(child, level + 1)

def findNode(node, value):
    if not node: return None
    if node.value == value: return node
    for i in node.children:
        res = findNode(i, value)
        if res: return res

def findNodeDfsStack(rt, value):
    if rt is None: return None
    stack = [rt]
    while stack:
        node = stack.pop()
        if node.value == value:
            return node
        for i in node.children:
            stack.append(i)
```

```python
from collections import deque
p = TreeNode(1)
N = 127
q = deque([p])
cnt = 1
```

```python
while cnt < N:
    node = q.popleft()
    leftChild = TreeNode(node.value * 2)
    rightChild = TreeNode(node.value * 2 + 1)

    node.addChild(leftChild)
    node.addChild(rightChild)
    q.append(leftChild)
    q.append(rightChild)
    
    cnt += 2

p.children[1].children[1].children[1].children[1].children[1].children[1].value
```

```python
ntf = findNode(p, 16)
ntfDfs = findNodeDfsStack(p, 16)
childrenVals = [ i.value for i in ntf.children]
print(f"value - {ntf.value}\nchildren - {childrenVals}")
print("------- DFS ALG -------")
childrenValsDfs = [ i.value for i in ntfDfs.children]
print(f"value - {ntf.value}\nchildren - {childrenValsDfs}")
```
